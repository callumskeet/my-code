# Lemon Count

```python
import numpy as np

from sklearn.cluster import DBSCAN, KMeans

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
images = ['lemon_count.jpg', 'lemon_count2.jpg']

for image in images:
    # show original image
    lemon_img = mpimg.imread(image)
    plt.imshow(lemon_img)
    plt.title("%s - Original" % image)
    plt.show()
    
    # prepare clustering vars
    rgb_shape = lemon_img.shape    # store image dim and rgb layout
    cluster_shape = [rgb_shape[0]*rgb_shape[1], 3]
    X = np.reshape(lemon_img, cluster_shape)

    # perform K-Means clustering to group the pixels into K clusters
    # clusters are formed based on pixel's location and RGB value
    K = 6
    km = KMeans(n_clusters=K, init='k-means++')
    km.fit(X)

    # recolour pixels of each cluster with the centroid colour value
    clusters = len(set(km.labels_))
    for k in range(clusters):
        member_mask = (km.labels_ == k)
        X[member_mask,:] = km.cluster_centers_[k,:]

    # show recoloured image
    cluster_clr_img = np.reshape(X, rgb_shape)
    plt.imshow(cluster_clr_img)
    plt.title("K-Means - # of clusters = %d" % clusters)
    plt.show()

    # find which cluster contains the lemons
    yellow = [255, 255, 0]    # RGB representation of yellow
    min_dist = 999999999.0
    lemon_cluster = 0
    for k in range(clusters):
        dist = np.linalg.norm(yellow - km.cluster_centers_[k,:])
        if (dist < min_dist):
            min_dist = dist
            lemon_cluster = k

    # Remove irrelevant data - ie clusters that are not the lemons
    binary_array = np.zeros_like(X)  # 0 for noise, 1 for lemon cluster member
    for k in range(clusters):
        mask = (km.labels_ == k)
        if k == lemon_cluster:
            X[mask,:] = [255, 255, 255] # white
            binary_array[mask,:] = 1
        else:
            X[mask,:] = [0, 0, 0] # black (noise)
    binary_array = binary_array[:, 0]

    # show members of the lemon cluster
    lemons_bw = np.reshape(X, rgb_shape)
    plt.imshow(lemons_bw)
    plt.title("Pixels part of lemon cluster (k = %d)" % lemon_cluster)
    plt.show()

    # create an array containing the coordinates of each cluster member
    # reducing the complexity of the data greatly improves the performance of DBSCAN
    binary_matrix = np.reshape(binary_array, [rgb_shape[0], rgb_shape[1]])
    coords = []
    row_no = 0
    for row in binary_matrix:
        col_no = 0
        for pos in row:
            if pos == 1:
                coords.append([col_no, row_no])
            col_no += 1
        row_no += 1
    coords = np.array(coords)

    # plot the simplified data
    plt.scatter(coords[:,0], coords[:,1])
    plt.gca().invert_yaxis()
    plt.title("Scatter plot created from coordinates of lemon cluster members")
    plt.show()

    # perform DBSCAN on simplified data
    db = DBSCAN(eps=4, min_samples=10).fit(coords)
    lemon_count = len(set(db.labels_))

    colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])

    for k in range(lemon_count):
        k_col = colors[k]
        mask = (db.labels_ == k)
        
        xy = coords[mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=k_col, markeredgecolor=k_col)
        
    plt.title("DBSCAN - number of lemons = %d" % lemon_count)
    plt.gca().invert_yaxis()
    plt.show()
```
